```python
import turtle
import tkinter as tk
from tkinter import messagebox as msg
from turtle import TurtleScreen,RawTurtle
#creating gui
win = tk.Tk()
win.geometry("800x600")
win.resizable(width = False,height = False)
win.title("----Turtle Race----")
win["bg"] = "green"
canvas = tk.Canvas(win, width = 750, height = 455)
canvas.place(x = 20,y = 10)
title = tk.Label(canvas,text = "Turtle Race",width = 20,height = 1,font = ("Agency FB",40))
title.place(x = 200,y = 10)

chance =""
def start():
    #Turtle one for player one
    
    s1 = TurtleScreen(canvas)
    s1.bgcolor("white")
    player_one = RawTurtle(s1)
    
    player_one.color("red")  #set up color for turtle 1
    player_one.shape("turtle") #Shape of the turtle
    player_one.penup()  # penup for starting position
    player_one.goto(-200,100)  #set position for turtle

    #for player two clone the turtle
    player_two = player_one.clone() 
    player_two.color("black") #change the color of the turtle
    player_two.penup()  #for starting position
    player_two.goto(-200,-100) #position for second turtle

    #setup victory line for both tutrle
    player_one.goto(300,60) #on selected position
    player_one.pendown() #then pendown and mark location
    player_one.circle(40)   #Circle for 1st player
    player_one.penup()
    player_one.goto(-200,100)  #come back at initial position

    player_two.goto(300,-140) #on selected position for second
    player_two.pendown() # pendown
    player_two.circle(40) #mark victory line
    player_two.penup()
    player_two.goto(-200,-100) #come back
    
    import random
    #creating a dice
    die = [1,2,3,4,5,6]
    def play():
        for i in range(20):
            global chance
            if chance != "A":
                chance = "A"
                if player_one.pos() >= (300,50):
                        print("Player One Wins!")
                        win.destroy()
                        win2 = tk.Tk()
                        
                        win2.title("Congratulations")
                        win2.geometry("400x100")
                        win2["bg"] = "yellow"
                        tk.Label(win2,text = "Player One win this game",font = ("Agency FB",20)).pack()
                        
                        
                        break
                
                else:
                        
                        die_outcome = random.choice(die)
                        #print("The result of the die roll is: ")
                        #print(die_outcome)
                        #print("The number of steps will be: ")
                        #print(20*die_outcome)
                        player_one.fd(20*die_outcome)
                        print("player one position ===>",player_one.pos())
            else:
                    msg.showwarning("Alert","It not your chance")
            
            break

     #player two button working       
    def play2():
        for i in range(20):
            global chance
            if chance !="B":
                chance = "B"
                

                if player_two.pos() >= (300,-50):
                    #msg.showinfo("Congrates","Player 2 You win this Game")
                    print("Player Two Wins!")
                    win.destroy()
                    win2 = tk.Tk()
                    win2.title("Congratulations")
                    win2.geometry("400x200")
                    win2["bg"] = "yellow"
                    tk.Label(win2,text = "Player Two win this game",font = ("Agency FB",20)).pack()
                    break
                else:
                    die_outcome2 = random.choice(die)
                    
                    #print("The result of the die roll is: ")
                    #print(die_outcome2)
                    #print("The number of steps will be: ")
                    #print(20*die_outcome2)
                    player_two.fd(20*die_outcome2)
                    print("player two position ===>",player_two.pos())

            else:
                msg.showwarning("Alert","It not your chance")
            
            break
    player1 = tk.Button(canvas,text = "Player 1",bg = "green",fg = "white",
                         activeforeground = "yellow",
                         font = ("Agency FB",18),
                         relief = "ridge",width = 8,height = 1,command = play)
    player1.place(x = 230, y = 420)
    player2 = tk.Button(canvas,text = "Player 2",bg = "green",fg = "white",
                         activeforeground = "yellow",
                         font = ("Agency FB",18),
                         relief = "ridge",width = 8,height = 1,command = play2)
    player2.place(x = 500, y = 420)
    
start_button = tk.Button(canvas,text = "Start",bg = "green",fg = "white",
                         activeforeground = "yellow",
                         font = ("Agency FB",18),command = start,
                         relief = "ridge",width = 8,height = 1)
start_button.place(x = 200,y =20)
stop_button = tk.Button(canvas,text = "Stop",bg = "green",fg = "white",
                         activeforeground = "yellow",
                         font = ("Agency FB",18),command = win.destroy,
                         relief = "ridge",width = 8,height = 1)
stop_button.place(x = 500,y =20)

win.mainloop()
```
